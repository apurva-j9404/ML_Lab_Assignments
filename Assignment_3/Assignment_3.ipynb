```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Admission_Predict.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GRE Score',y='Chance of Admit ',hue='University Rating',data=df ,palette="coolwarm",s=100)
plt.title("Scatter Plot: GRE Score vs Chance of Admit")
plt.xlabel("GRE Score")
plt.ylabel("Chance of Admit")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=df['University Rating'],y=df['CGPA'],palette="viridis")
plt.title("Bar Plot: Average CGPA by University Rating")
plt.xlabel("University Rating")
plt.ylabel("Average CGPA")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Research'],y=df['TOEFL Score'],palette="pastel")
plt.title("Box Plot: TOEFL Score by Research")
plt.xlabel("Research (0 = No, 1 = Yes)")
plt.ylabel("TOEFL Score")
plt.show()
```

```python
research_counts = df['Research'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(research_counts, labels=['No Research', 'Has Research'], autopct='%1.1f%%', colors=['red', 'green'], startangle=90)
plt.title("Pie Chart: Research vs No Research Applicants")
plt.show()
```

```python
df_sorted = df.sort_values(by="TOEFL Score")
plt.figure(figsize=(6,4))
plt.plot(df_sorted['Serial No.'], df_sorted['TOEFL Score'], marker='o', linestyle='-', color='blue')
plt.title("Line Chart: TOEFL Score Trend")
plt.xlabel("Application Serial No.")
plt.ylabel("TOEFL Score")
plt.show()
```
